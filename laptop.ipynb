```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\bhara\Downloads\laptop_price.csv", encoding = 'latin1')
df.head()
```

```python
df.drop(columns = {'laptop_ID'}, inplace = True, axis = 1)
```

```python
df.info()
```

```python
df['Ram'] = df['Ram'].str.split('GB').str.join('').astype(int)
df['Ram'].info()
```

```python
df['Weight'] = df['Weight'].str.split('kg').str.join('')
df['Weight'] = df['Weight'].str.replace('kg', '', regex=False).astype(float)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X = df.drop('Price_euros', axis = 1)
Y = df['Price_euros']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
df.head(1)
```

```python
ohe = OneHotEncoder()
cat_cols = ['Company', 'Product', 'TypeName', 'ScreenResolution', 'Cpu', 'Memory', 'Gpu', 'OpSys']
ohe.fit(X[cat_cols])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import BaggingRegressor, AdaBoostRegressor, VotingRegressor, RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
```

```python
lr = LinearRegression()
dt = DecisionTreeRegressor()
bg = BaggingRegressor()
abr = AdaBoostRegressor()
rf = RandomForestRegressor()
gbr = GradientBoostingRegressor()
svr = SVR()
xg = XGBRegressor()
```

```python
vc = VotingRegressor(estimators = [('lr', lr), ('dt', dt), ('bg', bg), ('abr', abr), ('rf', rf), ('gbr', gbr), ('svr', svr), ('xg', xg)])
```

```python
X_train.head()
```

```python
trf = ColumnTransformer(transformers = [
    ('ohe', OneHotEncoder(categories = ohe.categories_), ['Company', 'Product', 'TypeName', 'ScreenResolution', 'Cpu', 'Memory', 'Gpu', 'OpSys'])
], remainder = 'passthrough')
```

```python
pipe = make_pipeline(trf, xg)
pipe.fit(X_train, Y_train)
r2_score(Y_test, pipe.predict(X_test))
```

```python
import pickle
with open('model.pkl', 'wb') as f:
    pickle.dump(pipe, f)
```
